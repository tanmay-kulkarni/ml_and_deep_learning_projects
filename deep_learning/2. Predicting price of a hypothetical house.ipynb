# Another example

In this notebook, we'll try to predict the price of a house which follows a simple pricing rule. In reality, the cost of a house depends on a lot of factors, but let's just keep it simple for the time being. We begin by importing some standard packages.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# House price

Let's consider a simple pricing scheme.

    Cost = 50k + (number_of_bedrooms * 50k)

Thus, cost of a one bedroom house is 100k, cost of a two bedroom house is 150k and so on. Let's generate some data that satisfies this rule.

```python
X = [1, 2, 3, 4, 5, 6]
y = [50000 + x*50000 for x in X]

for t in list(zip(X,y)):
    print('#bedrooms: {} | cost of house: ${}'.format(t[0], t[1]))
```

# Create a neural network model

Again, let's go with the same model which I used in the previous notebook. i.e. a single neuron with 1 input and 1 output.

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

# Compile the model

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
```

# Fit the model

More the number of epochs, better will be the result. Notice how the loss keeps decreasing with each epoch.

```python
model.fit(X,y, epochs = 3000, verbose=True)
```

# Test the model

Let's predict the price of a house having 7 bedrooms

```python
model.predict([7])
```

# The actual price

We can find the actual price, since we know what rule is used to calculate it.

    Cost = 50k + (number_of_bedrooms * 50k)
    

```python
50000 + (7 * 50000)
```

With just 6 datapoints, our model was able to make a pretty good guess. The actual price of the house with 7 bedrooms is so close to the predicted price. Fascinating, isn't it?
